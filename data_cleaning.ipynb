```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("airport_traffic_2025.csv")
```

```python
df.size
```

```python
df.columns
```

```python
df = df.dropna(subset=['YEAR', 'MONTH_MON', 'FLT_DATE', 'APT_ICAO', 'APT_NAME', 'STATE_NAME',
       'FLT_DEP_1', 'FLT_ARR_1', 'FLT_TOT_1'])
```

```python
df[['FLT_DEP_IFR_2', 'FLT_ARR_IFR_2',
       'FLT_TOT_IFR_2']] = df[['FLT_DEP_IFR_2', 'FLT_ARR_IFR_2',
       'FLT_TOT_IFR_2']].fillna(0)
```

```python
df
```

```python
df.columns
```

```python
df = df.rename(columns={
    'YEAR': 'year',
    'MONTH_MON': 'month',
    'FLT_DATE': 'flight_date',
    'APT_ICAO': 'airport_icao',
    'APT_NAME': 'airport_name',
    'STATE_NAME': 'state',
    'FLT_DEP_1': 'departures',
    'FLT_ARR_1': 'arrivals',
    'FLT_TOT_1': 'total_flights',
    'FLT_DEP_IFR_2': 'ifr_departures',
    'FLT_ARR_IFR_2': 'ifr_arrivals',
    'FLT_TOT_IFR_2': 'ifr_total_flights'
})
```

```python
df.columns
```

```python
plt.plot(df['month'], df['total_flights'])
plt.xlabel('Month')
plt.ylabel('Total Flights')
plt.title('Monthly Flights Trend')
plt.show()
```

```python
df.to_csv('cleaned_airport-traffic_data.csv')
```
